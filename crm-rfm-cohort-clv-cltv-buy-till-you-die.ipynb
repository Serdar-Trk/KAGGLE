```python
#Modelimizin daha hizli calismasi adina hafiza kullaniminda iyilestirme yapmak istiyoruz.
#Veri setimizin hafiza kullanimina goz atalim.
memory_usage = df.memory_usage(deep=True) / 1024 ** 2
print('memory usage of features: \n', memory_usage)
print('memory usage sum: ',memory_usage.sum())
```

```python
# Veri turlerinde yapacagimiz degisiklikler ile hafiza kullanimini iyilestirelim.
# Bu sayede hafiza kullanimini %73 oraninda azalttik
def reduce_memory_usage(df, verbose=True):
    numerics = ["int8", "int16", "int32", "int64", "float16", "float32", "float64"]
    start_mem = df.memory_usage().sum() / 1024 ** 2
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == "int":
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                if (
                    c_min > np.finfo(np.float16).min
                    and c_max < np.finfo(np.float16).max
                ):
                    df[col] = df[col].astype(np.float16)
                elif (
                    c_min > np.finfo(np.float32).min
                    and c_max < np.finfo(np.float32).max
                ):
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
    end_mem = df.memory_usage().sum() / 1024 ** 2
    if verbose:
        print(
            "Mem. usage decreased to {:.2f} Mb ({:.1f}% reduction)".format(
                end_mem, 100 * (start_mem - end_mem) / start_mem
            )
        )
    return df

df = reduce_memory_usage(df, verbose=True)
```

```python
# Gormek istemedigimiz aykiri degerleri daha sonraki islemler icin temizleyelim

def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.15)
    quartile3 = dataframe[variable].quantile(0.85)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit.round(), up_limit.round()


def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
    
replace_with_thresholds(df, "Monatery")
replace_with_thresholds(df, 'NumWebPurchases') 
replace_with_thresholds(df, 'NumCatalogPurchases')
replace_with_thresholds(df, 'NumStorePurchases')
```

* Recency (R) : G√ºncellik. M√º≈üterinin son satƒ±n alma tarihinden bu yana ge√ßen zamanƒ± ifade eder. Daha yakƒ±n zamanda alƒ±≈üveri≈ü yapmƒ±≈ü m√º≈üterilerin, √ºr√ºnlerimizi kullanƒ±yor olduƒüunu, yakƒ±n zamanda bizimle temas kurduƒüunu, firmamƒ±zƒ± tekrar tercih etme ihtimalinin y√ºksek olduƒüunu d√º≈ü√ºnd√ºr√ºr. Dolayƒ±sƒ±yla Recency‚Äônin d√º≈ü√ºk olmasƒ±, yani son satƒ±n alma tarihinin yakƒ±n olmasƒ±, daha y√ºksek skorlarƒ± netice verecektir.

* Frequency (F) : Sƒ±klƒ±k. M√º≈üterinin satƒ±n alma sayƒ±sƒ±dƒ±r, basit√ße fatura sayƒ±sƒ± olarak d√º≈ü√ºn√ºlebilir. M√º≈üteri firmamƒ±zƒ± ne kadar √ßok sayƒ±da tercih etmi≈üse, gelecekte bu davranƒ±≈üƒ±nƒ± s√ºrd√ºrmesini bekleyebiliriz.

* Monetary (M) : Parasallƒ±k. M√º≈üterinin satƒ±n almalarƒ±nda ne kadar para √∂dediƒüini ifade eder. Frequency‚Äôye benzer ≈üekilde √∂denen para miktarƒ±, gruplandƒ±rmada √ºst segmentlerde yer almak konusunda pozitif etki eder.

```python
#Metriklerimizi olusturalim
df['Frequency'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']
RFM = df.loc[:, ['ID', 'Recency', 'Frequency', 'Monatery']]
RFM['Recency'].astype('int64')
RFM['Frequency'] = RFM['Frequency'].astype('int64')
RFM['Monatery'] = RFM['Monatery'].astype('int64')
RFM.head().style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

```python
print(" DESCRIBE ".center(70,'-'))
RFM.describe().style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

```python
# Belirledigimiz reciency_score ve frequency_score lari araciligiyla RFM skorumuzu belirleyelim
def get_rfm_scores(dataframe):

    dataframe["recency_score"] = pd.qcut(dataframe["Recency"], 5, labels=[5, 4, 3, 2, 1])
    dataframe["frequency_score"] = pd.qcut(dataframe["Frequency"], 5, labels=[1, 2, 3, 4, 5])
    dataframe["monetary_score"] = pd.qcut(dataframe["Monatery"], 5, labels=[1, 2, 3, 4, 5])
    dataframe["RFM_SCORE"] = dataframe["recency_score"].astype(str) + dataframe["frequency_score"].astype(str)

    return dataframe

get_rfm_scores(RFM)
RFM = RFM.reset_index()
RFM.head().style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

<div <h1 <strong style="color:#377D71;font-size:50px;font-family:Georgia;text-align:center;">CREATING & ANALYISING RFM SEGMENTS</strong></div> 

* Segment----------------------------->Description
* Champions-------------------------->Bought recently, buy often and spend the most
* Loyal Customers-------------------->Buy on a regular basis. Responsive to promotions.
* Potential Loyalist------------------->Recent customers with average frequency.
* Recent Customers------------------->Bought most recently, but not often.
* Promising---------------------------->Recent shoppers, but haven‚Äôt spent much.
* Customers Needing Attention------>Above average recency, frequency and monetary values. May not have bought very recently though.
* About To Sleep---------------------->Below average recency and frequency. Will lose them if not reactivated.
* At Risk------------------------------->Purchased often but a long time ago. Need to bring them back!
* Can‚Äôt Lose Them-------------------->Used to purchase frequently but haven‚Äôt returned for a long time.
* Hibernating------------------------->Last purchase was long back and low number of orders. May be lost.

<div style="float:center;margin-left:10px;max-width:1000%;">
    
<img src="https://serving.photos.photobox.com/11336762b8219b7dd1324921e832518251e56e818f63cc543e5284a27a3adf84fbb7b59f.jpg"></div>

* Ortalama Sipari≈ü Deƒüeri = Toplam Fiyat / Toplam ƒ∞≈ülem
* Satƒ±n Alma Sƒ±klƒ±ƒüƒ± = Toplam ƒ∞≈ülem / Toplam M√º≈üteri Sayƒ±sƒ±
* M√º≈üteri Deƒüeri = Ortalama Sipari≈ü Deƒüeri * Satƒ±n Alma Sƒ±klƒ±ƒüƒ±
* M√º≈üteri Terk Oranƒ± = 1 ‚Äî Tekrarlanan Sipari≈ü Oranƒ± [Birden fazla alƒ±≈üveri≈ü yapan m√º≈üteri sayƒ±sƒ± / T√ºm m√º≈üteriler]
* Kar Marjƒ± = Toplam Fiyat * (0,10) [0.10 deƒüeri √∂rnek olarak se√ßilmi≈ütir. ≈ûirketin stratejisine g√∂re deƒüi≈ükenlik g√∂sterir]
* CLTV = (M√º≈üteri Deƒüeri/M√º≈üteri Terk Oranƒ±) x Kar Marjƒ±
* M√º≈üteri ya≈üam boyu deƒüeri hesaplama y√∂ntemi yukarƒ±daki form√ºlasyon ile genel olarak ifade edilebilir.




Formula

Probabilistic lifetime value estimation with time projection

* average_order_value = total_price / total_transaction
* Purchase Frequency (total_transaction / total_number_of_customers)
* Repeat Rate & Churn Rate
* profit_margin =  total_price * 0.10
* customer_value = average_order_value * purchase_frequency)
* CLTV = (customer_value / churn_rate) x profit_margin)
* * Segmentation
CLTV = Expected Number of Transaction(BG/NBD) * Expected Average Profit(gamma gamma)

Above, purchase frequency and number of transaction mean the same thing. Likewise, Average Order Value and Average Profit mean the same thing. It differs with the Expected part that happens to them.

CAUTION: It will add a probabilistic distribution. Expected statement refers to this part. Expected number of purchases, expected profitability.

BG / NBD = Expected Transaction Gamma Gamma = Expected Profit

* recency: the difference between the customer's last purchase and his first purchase
* T: the age of the client in the company
* frequency: total number of repeat purchases
* monetary_value: average earnings per purchase

```python
# CLTV metriklerinin hesaplanmasƒ±
cltv_df = pd.DataFrame()
cltv_df["average_order_value"] = RFM["Monatery"] / RFM["Recency"]
cltv_df["purchase_frequency"] = RFM["Recency"] / RFM.shape[0]
repeat_rate = RFM[RFM["Recency"] > 1].shape[0] / RFM.shape[0]
churn_rate = 1 - repeat_rate
cltv_df['profit_margin'] = RFM['Monatery'] * 0.10
cltv_df['customer_value'] = cltv_df['average_order_value'] * cltv_df["purchase_frequency"]
cltv_df["cltv"] = (cltv_df["customer_value"] / churn_rate) * cltv_df["profit_margin"]

cltv_df.head().style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

```python
cltv_df.sort_values(by="cltv", ascending=False).head().style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

```python
# CLTV segmentlerinin belirlenmesi
cltv_df["segment"] = pd.qcut(cltv_df["cltv"], 4, labels=["D", "C", "B", "A"])

# Segmentlere g√∂re ayrƒ±ntƒ±lƒ± analizler yapabiliriz.

cltv_df.groupby("segment").agg({"count", "mean", "sum"}).style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

*  Prediction modellerimizi kurmadan √∂nce modelimiz de kullanacaƒüƒ±mƒ±z deƒüi≈ükenleri hesaplayalƒ±m. Bunlar:
*  'recency'  : M√º≈üterinin son alƒ±≈üveri≈üinin √ºzerinden ge√ßen zaman (haftalƒ±k)
*      'T'    :  M√º≈üterinin ya≈üƒ±(ilk alƒ±≈üveri≈üinin √ºzerinden ge√ßen zaman)
*  'frequency' : M√º≈üterinin alƒ±≈üveri≈ü sayƒ±sƒ±
*  'monetary'  : M√º≈üterinin yaptƒ±ƒüƒ± alƒ±≈üveri≈ülerin toplam tutarƒ±

```python
cltv = pd.DataFrame()

cltv["customer_id"] = df["ID"]
cltv["recency_cltv_daily"] = (df['Recency'])
cltv["T_daily"] = (df['Dt_Customer'].max() - df['Dt_Customer']).dt.days + 86
cltv["T_daily"]
cltv["monetary_cltv_avg"] = (df["Monatery"] / df["Frequency"]).round()

cltv["recency_cltv_daily"] = cltv["recency_cltv_daily"].astype(int)
cltv["T_daily"] = cltv["T_daily"].astype(int)
cltv["Frequency"] = df["Frequency"].astype(int)

cltv = cltv[(cltv['Frequency'] > 1)]
cltv.head().style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

```python
#BGF Modelimizi kuralƒ±m

bgf = BetaGeoFitter(penalizer_coef=0.001)

bgf.fit(cltv['Frequency'],
        cltv['recency_cltv_daily'],
        cltv['T_daily'])
```

```python
# 1 aylƒ±k s√ºre√ßte alƒ±≈üveri≈ü yapma ihtimali en y√ºksek olan m√º≈üteriler.

cltv["expected_purc_1_month"] = bgf.conditional_expected_number_of_purchases_up_to_time(30,
                                                        cltv['Frequency'],
                                                        cltv['recency_cltv_daily'],
                                                        cltv['T_daily'])
cltv.head().style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

```python
# 3 aylƒ±k s√ºre√ßte alƒ±≈üveri≈ü yapma ihtimali en y√ºksek olan m√º≈üteriler.

cltv["expected_purc_3_month"] = bgf.predict(3 * 30,
                                               cltv['Frequency'],
                                               cltv['recency_cltv_daily'],
                                               cltv['T_daily'])
cltv.head(10).style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

```python
# Gamma gamma modeli
ggf = GammaGammaFitter(penalizer_coef=0.01)
```

```python
ggf.fit(cltv['Frequency'], cltv['monetary_cltv_avg'])
```

```python
ggf.conditional_expected_average_profit(cltv['Frequency'],
                                        cltv['monetary_cltv_avg']).sort_values(ascending=False).head(10)
```

```python
cltv["expected_average_profit"] = ggf.conditional_expected_average_profit(cltv['Frequency'],
                                                                             cltv['monetary_cltv_avg'])
cltv.head().style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

```python
cltv.sort_values("expected_average_profit", ascending=False).head(10).style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

```python
cltv_result = pd.DataFrame()
cltv_result["customer_id"] = cltv["customer_id"]
cltv_result['clv'] = ggf.customer_lifetime_value(bgf,
                                   cltv['Frequency'],
                                   cltv['recency_cltv_daily'],
                                   cltv['T_daily'],
                                   cltv['monetary_cltv_avg'],
                                   time=365,  # 1 yillik
                                   freq="D",  # T'nin frekans bilgisi.
                                   discount_rate=0.001)
cltv_result.head().sort_values(by='clv', ascending=False)
```

```python
# Olusturdugumuz CLV degerini datafram e ekleyelƒ±m.
cltv_result = cltv_result.reset_index()
clv = cltv.merge(cltv_result, on="customer_id", how="left")
clv.head(10).style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```
