```python
def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.01)
    quartile3 = dataframe[variable].quantile(0.80)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit


def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
    
replace_with_thresholds(df, "quantity")
replace_with_thresholds(df, "amount")

print(" DESCRIBE ".center(90,'~'))
print(df.describe().T)
```

```python
# Let's calculate the RFM metrics..
# Recency : The time elapsed since the customer's last purchase.
# Frequency: The total number of purchases by the customer.
# Monetary: The total money left by the customer in the whole process.


rfm = df.groupby('customerID').agg({'invoice_date': lambda x: (today_date - x.max()).days,
                                     'invoiceID': lambda x: x.nunique(),
                                     'amount': lambda x: x.sum()})

rfm.columns = ['recency', 'frequency', 'monetary']
# Ä°adelerden doÄŸabilecek negatif deÄŸerleri ve 0 deÄŸerleri kaldÄ±ralÄ±m.
rfm = rfm[rfm["monetary"] > 0]

b=rfm.head()

table = ff.create_table(b,colorscale="blackbody")
for i in range(len(table.layout.annotations)):
    table.layout.annotations[i].font.size = 13
iplot(table)
```

```python
# Let's determine the RFM scores with the function we will create.

def get_rfm_scores(dataframe):

    dataframe["recency_score"] = pd.qcut(dataframe["recency"], 5, labels=[5, 4, 3, 2, 1])
    dataframe["frequency_score"] = pd.qcut(dataframe["frequency"].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
    dataframe["monetary_score"] = pd.qcut(dataframe["monetary"], 5, labels=[1, 2, 3, 4, 5])
    dataframe["RFM_SCORE"] = dataframe["recency_score"].astype(str) + dataframe["frequency_score"].astype(str)

    return dataframe

get_rfm_scores(rfm)
rfm.reset_index()
rfm.head()
```

```python
# In Cohort Analysis, we track groups of users over time who find some common patterns or behaviors.
def CohortAnalysis(dataframe):
    dataframe = dataframe[["customerID", "invoiceID", "invoice_date"]].drop_duplicates()
    dataframe["order_month"] = dataframe["invoice_date"].dt.to_period("M")
    dataframe["cohort"] = (dataframe.groupby("customerID")["invoice_date"].transform("min").dt.to_period("M"))
    cohort_data = (dataframe.groupby(["cohort", "order_month"]).agg(n_customers=("customerID", "nunique")).reset_index(drop=False))
    cohort_data["period_number"] = (cohort_data.order_month - cohort_data.cohort).apply(attrgetter("n"))
    cohort_pivot = cohort_data.pivot_table(index="cohort", columns="period_number", values="n_customers")
    cohort_size = cohort_pivot.iloc[:, 0]
    retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
    with sns.axes_style("white"):
        fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={"width_ratios": [1, 11]})
        sns.heatmap(retention_matrix,mask=retention_matrix.isnull(), annot=True, cbar=False, fmt=".0%", cmap="inferno", ax=ax[1],)
        ax[1].set_title("Monthly Cohorts: User Retention", fontsize=14)
        ax[1].set(xlabel="# of periods", ylabel="")
        white_cmap = mcolors.ListedColormap(["white"])
        sns.heatmap(pd.DataFrame(cohort_size).rename(columns={0: "cohort_size"}),annot=True,cbar=False,fmt="g",cmap=white_cmap,ax=ax[0],)
        fig.tight_layout()
    
CohortAnalysis(df)
```

<a id = "8"></a>
<p style= "background-color:#3F497F;font-family:Georgia;color:#FEFBE9;font-size:180%;text-align:center;border-radius:20px 60px;overflow:hidden;border-style:dotted;border-width:1.8px;border-color:#FEFBE9;"><b>CUSTOMER LIFETIME VALUE</b></p>

* average_order_value = total_price / total_transaction)
* Purchase Frequency (total_transaction / total_number_of_customers)
* Repeat Rate & Churn Rate
* profit_margin =  total_price * 0.10
* customer_value = average_order_value * purchase_frequency)
* CLTV = (customer_value / churn_rate) x profit_margin)
* Segmentation
* CLTV = Expected Number of Transaction(BG/NBD) * Expected Average Profit(gamma gamma)

```python
# 'total_transaction' : Number of unique invoices per customer
# 'total_unit'        : Total number of products purchased specifically for the customer
# 'total_price'        :  Total amount spent per customer

clv = df.groupby('customerID').agg({'invoiceID': lambda x: x.nunique(),
                                        'quantity': lambda x: x.sum(),
                                        'amount': lambda x: x.sum()})

clv.columns = ['total_transaction', 'total_unit', 'total_price']

clv['total_transaction'] = clv['total_transaction'].astype(float)
clv.reset_index()
d=clv.head()

table = ff.create_table(d,colorscale="blackbody")
for i in range(len(table.layout.annotations)):
    table.layout.annotations[i].font.size = 13
iplot(table)
```

```python
# Calculating CLV metrics
clv["average_order_value"] = clv["total_price"] / clv["total_transaction"]
clv["purchase_frequency"] = clv["total_transaction"] / clv.shape[0]
repeat_rate = clv[clv["total_transaction"] > 1].shape[0] / clv.shape[0]
churn_rate = 1 - repeat_rate
clv['profit_margin'] = clv['total_price'] * 0.33
clv['customer_value'] = clv['average_order_value'] * clv["purchase_frequency"]

clv["clv"] = (clv["customer_value"] / churn_rate) * clv["profit_margin"]
clv.head()
```

```python
# SÄ±ralÄ± verimize bir gÃ¶z atalÄ±m.
clv.sort_values(by="clv", ascending=False).head()
```

```python
# CLV segmentlerinin belirlenmesi
clv["segment"] = pd.qcut(clv["clv"], 4, labels=["D", "C", "B", "A"])

# Segmentlere gÃ¶re ayrÄ±ntÄ±lÄ± analizler yapabiliriz.

clv.groupby("segment").agg({"count", "mean", "sum"})
```

```python
# Before setting up our prediction models, let's calculate the variables that you will use in our model. These:
# 'recency'  : Time since customer's last purchase (weekly)
#     'T'    :  Customer's age (time since first purchase)
# 'frequency' : Number of purchases by the customer
#  'monetary'  : The total amount of purchases made by the customer.

clv_predict = df.groupby('customerID').agg({'invoice_date': [lambda InvoiceDate: (InvoiceDate.max() - InvoiceDate.min()).days,
                                                         lambda InvoiceDate: (today_date - InvoiceDate.min()).days],
                                         'invoiceID': lambda Invoice: Invoice.nunique(),
                                         'amount': lambda TotalPrice: TotalPrice.sum()})

clv_predict.columns = ['recency', 'T', 'frequency', 'monetary']

clv_predict["monetary"] = clv_predict["monetary"] / clv_predict["frequency"]

clv_predict = clv_predict[(clv_predict['frequency'] > 1)]
clv_predict["recency"] = clv_predict["recency"] / 7
clv_predict["T"] = clv_predict["T"] / 7
clv_predict.reset_index()
d=clv_predict.head()

table = ff.create_table(d,colorscale="blackbody")
for i in range(len(table.layout.annotations)):
    table.layout.annotations[i].font.size = 13
iplot(table)
```

```python
#BGF Modelimizi kuralÄ±m

bgf = BetaGeoFitter(penalizer_coef=0.001)

bgf.fit(clv_predict['frequency'],
        clv_predict['recency'],
        clv_predict['T'])
```

```python
# Customers most likely to make a purchase in the 1-month period.
clv_predict["expected_purc_1_month"] = bgf.conditional_expected_number_of_purchases_up_to_time(4,
                                                        clv_predict['frequency'],
                                                        clv_predict['recency'],
                                                        clv_predict['T'])
clv_predict.head()
```

```python
# Customers who are most likely to make a purchase in the 3-month period.

clv_predict["expected_purc_3_month"] = bgf.predict(4 * 3,
                                               clv_predict['frequency'],
                                               clv_predict['recency'],
                                               clv_predict['T'])
clv_predict.head(10)
```

```python
# Gamma gamma model
ggf = GammaGammaFitter(penalizer_coef=0.01)
```

```python
# setup of our gamma gamma model 
ggf.fit(clv_predict['frequency'], clv_predict['monetary'])
```

```python
# Calculation of expected average profit.

ggf.conditional_expected_average_profit(clv_predict['frequency'],
                                        clv_predict['monetary']).sort_values(ascending=False).head(10)
```

```python
clv_predict["expected_average_profit"] = ggf.conditional_expected_average_profit(clv_predict['frequency'],
                                                                             clv_predict['monetary'])
clv_predict.head()
```

```python
clv_predict.sort_values("expected_average_profit", ascending=False).head(10)
```

```python
# calculating CLTV(Customer Lifetime Value)

clv_result = ggf.customer_lifetime_value(bgf,
                                   clv_predict['frequency'],
                                   clv_predict['recency'],
                                   clv_predict['T'],
                                   clv_predict['monetary'],
                                   time=3,  # 3 aylÄ±k
                                   freq="W",  # T'nin frekans bilgisi.
                                   discount_rate=0.01)
clv_result.head().sort_values(ascending=False)
```

```python
# Let's add the CLTV value we created to the dataframe.
clv_result = clv_result.reset_index()
clv_final = clv_predict.merge(clv_result, on="customerID", how="left")

clv_final.head(10)
```

```python
clv_final.sort_values(by="clv", ascending=False).head(10)
```
