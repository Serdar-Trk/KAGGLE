```python
import pandas as pd
df = pd.read_csv('/kaggle/input/amazon-reviews/amazon_reviews.csv')
```

```python
def check_df(dataframe, head=5):
    print(" SHAPE ".center(70,'-'))
    print('Rows: {}'.format(dataframe.shape[0]))
    print('Columns: {}'.format(dataframe.shape[1]))
    print(" TYPES ".center(70,'-'))
    print(dataframe.dtypes)
    print(" MISSING VALUES ".center(70,'-'))
    print(dataframe.isnull().sum())
    print(" DUPLICATED VALUES ".center(70,'-'))
    print(dataframe.duplicated().sum())
    print(" DESCRIBE ".center(70,'-'))
    print(dataframe.describe())
check_df(df)
```

```python
# Let's change the data type of the comment variable to datetime.
df['reviewTime'] = pd.to_datetime(df['reviewTime'])
```

```python
df["helpful"].head()
```

```python
# Let's convert the variable to get the liked and disliked comments.
new_features = df["helpful"].str.split(",", expand=True)
new_features.dtypes
```

```python
new_features.head()
```

```python
#Let's create the variable df["helpful_yes"]
df["helpful_yes"] = new_features[0].str.lstrip("[")
df["helpful_yes"] = df["helpful_yes"].astype("int64")
df["helpful_yes"].head()
```

```python
#Let's create the variable df["total_vote"]
df["total_vote"] = new_features[1].str.rstrip("]")
df["total_vote"] = df["total_vote"].astype("int64")
df['total_vote'].head()
```

```python
#Let's create the variable df["helpful_no"]
df["helpful_no"] = df["total_vote"] - df["helpful_yes"]
df["helpful_no"].head()
```

 <div <h1      <strong style="color:#3795BD;font-size:40px;font-family:Georgia;text-align:center;">Rating Calculation</strong></div>



```python
# Let's observe in order according to Rating.
df.sort_values("overall", ascending=False).head()
```

```python
#The average of the rats will not be enough.
df['overall'].mean()
```

```python
#Look at their weighted averages by point times.
df['reviewTime'] = pd.to_datetime(df['reviewTime'])
current_date = dt.datetime.today()
df['days'] = (current_date - df['reviewTime']).dt.days
```

```python
# Let's determine the score times, it is possible to set it manually.
a = df['days'].quantile(0.25)
b = df['days'].quantile(0.50)
c = df['days'].quantile(0.75)
# The weights that we will determine in the weighted average may change according to the degree of importance.
def time_based_weighted_average(dataframe, w1=28, w2=26, w3=24, w4=22):
    return dataframe.loc[dataframe["days"] <= a, "overall"].mean() * w1 / 100 + \
           dataframe.loc[(dataframe["days"] > a) & (dataframe["days"] <= b), "overall"].mean() * w2 / 100 + \
           dataframe.loc[(dataframe["days"] > b) & (dataframe["days"] <= c), "overall"].mean() * w3 / 100 + \
           dataframe.loc[(dataframe["days"] > c), "overall"].mean() * w4 / 100

time_based_weighted_average(df)

# We observe that the change we have made in time weights is reflected in the overall score.
```

```python
# Let's execute the wilson_lower_bound function.
def wilson_lower_bound(pos, neg, confidence=0.95):
    import scipy.stats as st
    import math
    n = pos + neg
    if (pos - neg) == 0:
        return 0
    z = st.norm.ppf(1 - (1 - confidence) / 2)
    phat = 1.0 * pos / n
    return (phat + z * z / (2 * n) - z * math.sqrt((phat * (1 - phat) + z * z / (4 * n)) / n)) / (1 + z * z / n)
```

```python
#And we apply our wilson_lower_bound sorting function.
df["wilson_lower_bound"] = df.apply(lambda x: wilson_lower_bound(x["helpful_yes"], x["helpful_no"]), axis=1)
df.loc[:, ["wilson_lower_bound", 'reviewerID', 'reviewText']].sort_values(by='wilson_lower_bound',
                                                                          ascending=False).head(20)
```

```python
#Sorting by rating
df.sort_values("overall", ascending=True).head()
```

```python
#Sorting by Comment
df['comment_count'] = df['helpful_yes'] + df['helpful_yes']
df.sort_values(by="comment_count", ascending=False).head()
```

```python
# Let's weight it according to the average and the number of comments.
df["comment_count_scaled"] = MinMaxScaler(feature_range=(1, 5)).fit(df[["comment_count"]]).transform(df[["comment_count"]])

def weighted_sorting_score(dataframe, w1=55, w2=45):
    return (dataframe["comment_count_scaled"] * w1 / 100 +
            dataframe["overall"] * w2 / 100)

df["weighted_sorting_score"] = weighted_sorting_score(df)

df.sort_values("weighted_sorting_score", ascending=False).head()

```

<div <h1 <strong style="color:#3795BD;font-size:20px;font-family:Georgia;">  weighted_rating = (v/(v+M) * r) + (M/(v+M) * C) </strong></div>

* r = vote average
* v = vote count
* M = minimum votes required to be listed in the Top 250
* C = the mean vote across the whole report (currently 7.0)

```python
#Let's apply our formula to our variable with a score of 8.5 and a vote count of 8358 as follows.
def weighted_rating(r, v, M, C):
    return (v / (v + M) * r) + (M / (v + M) * C)

M = 2500
C=5.6

weighted_rating(8.50, 8358, M, C)
```

```python
# Observe by applying our weighted scores to our entire table.

movies["weighted_rating"] = weighted_rating(movies["vote_average"],
                                        movies["vote_count"], M, C)

movies.sort_values("weighted_rating", ascending=False).head(10).style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```

<div <h1 <strong style="color:#3795BD;font-size:20px;font-family:Georgia;">  Bayesian Average Rating Score </strong></div>

*  12481..........                                    The Dark Knight
*  314 ..........                            The Shawshank Redemption
*  2843 .........                                         Fight Club
*  15480  ......                                        Inception
*  292  .........                                       Pulp Fiction

```python
# confidence=0.95  --> % 95
def bayesian_average_rating(n, confidence=0.95):
    if sum(n) == 0:
        return 0
    K = len(n)
    z = st.norm.ppf(1 - (1 - confidence) / 2)
    N = sum(n)
    first_part = 0.0
    second_part = 0.0
    for k, n_k in enumerate(n):
        first_part += (k + 1) * (n[k] + 1) / (N + K)
        second_part += (k + 1) * (k + 1) * (n[k] + 1) / (N + K)
    score = first_part - z * math.sqrt((second_part - first_part * first_part) / (N + K + 1))
    return score


# bayesian_average_rating formulunun asagidaki aldigi degerler yildiz sayilarini ifade etmektedir.
# bayesian_average_rating(1_yildiz_sayisi, 2_yildiz_sayisi......)
# Asagidaki islemlerde daha iyi gozlemlenebilecektir

bayesian_average_rating([34733, 4355, 4704, 6561, 13515, 26183, 87368, 273082, 600260, 1295351])
```

```python
movies2 = pd.read_csv("/kaggle/input/imdb-ratings/imdb_ratings.csv")
movies2 = movies2.iloc[0:, 1:]
```

```python
movies2["bar_score"] = movies2.apply(lambda x: bayesian_average_rating(x[["one", "two", "three", "four", "five",
                                                                "six", "seven", "eight", "nine", "ten"]]), axis=1)
movies2.sort_values("bar_score", ascending=False).head(20).style.set_properties(**{"background-color": "#000000","color":"#FFEA20","border": "0.2px solid white"})
```
